```python
y=df["Performance Index"].values
x=df.drop("Performance Index" , axis=1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.12,random_state=1234)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train
```

```python
x_test
```

```python
 import tensorflow
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
tensorModel = Sequential()
```

```python
tensorModel.add(Dense(5,activation="relu"))
tensorModel.add(Dense(64,activation="relu"))
tensorModel.add(Dense(2,activation="relu"))
tensorModel.add(Dense(1))

tensorModel.compile(optimizer="adam" , loss="mse" , metrics=["mae"])
```

```python
tensorModel.fit(x=x_train,y=y_train , epochs=200,verbose=1,validation_data=(x_test,y_test))
```

```python
modelPlot=pd.DataFrame(tensorModel.history.history)
```

```python
tensorModel.summary()
```

```python
modelPlot.plot()
```

```python
tensorModel.evaluate(x_test,y_test)
```

```python
prediction=tensorModel.predict(x_test)
```

```python
prediction
```

```python
print(r2_score(prediction,y_test))
```

```python
print(mean_absolute_error(prediction,y_test))
print(mean_squared_error(prediction,y_test))
```


<div style="background-color: Green; padding: 16px; border: 2px white; color: white; font-size:29px;">
